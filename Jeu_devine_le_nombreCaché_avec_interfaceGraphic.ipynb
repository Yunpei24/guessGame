```python
import tkinter as tk
from tkinter import ttk
import random
import time

class GuessNumberGame(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Devine le nombre caché")
        self.geometry("680x560")
        self.configure(bg='#03F2F1')
        self.secret_number1 = random.randint(1, 100) #pour le niveau debutant
        self.secret_number2 = random.randint(1, 250) #pour le niveau moyen
        self.secret_number3 = random.randint(1, 1000) #pour le niveau difficile
        self.compteur = 0
        self.time_start1 = time.time()
        self.time_limit1 = 60
        self.time_start2 = time.time()
        self.time_limit2 = 45
        self.time_start3 = time.time()
        self.time_limit3 = 30
        
        self.niveau = ['Niveau1', 'Niveau2', 'Niveau3']
        self.guess_var = tk.StringVar()
        self.result_var = tk.StringVar()
        self.msg_niveau_var = tk.StringVar()
        self.result_var.set("BIENVENUE")
        
        self.niveau_frame = ttk.Frame(self, padding=10, relief='solid', borderwidth=1, style='My.TFrame')
        self.niveau_frame.pack(pady=22)
        self.niveau_label = ttk.Label(self.niveau_frame, text="Choisir le niveau", font=("Helvetica", 12, 'bold'), background='#f0f0f0')
        self.niveau_label.grid(row=0, column=0)
        self.choixNiveau = ttk.Combobox(self.niveau_frame , values = self.niveau, width=25)
        self.choixNiveau.grid(row=0, column=1)
        
        
        # Creating a style for the widgets
        s = ttk.Style()
        s.configure('My.TFrame', background='#03F2F1')
        s.configure('My.TLabel', background='#f0f0f0', font=("Helvetica", 14), foreground='black')
        s.configure('My.TButton', font=("Helvetica", 14), foreground='black', background='#ADD8E6')
        s.configure('My.TEntry', font=("Helvetica", 14), background='#F5F5DC')
        
        self.create_widgets()
    
    def create_widgets(self):
        title_frame = ttk.Frame(self, padding=10, relief='solid', borderwidth=1)
        title_frame.pack(pady=20)
        
        title_label = ttk.Label(title_frame, text="Devine le nombre caché", font=("Helvetica", 18, 'bold'), background='#f0f0f0')
        title_label.pack()
        
        guess_frame = ttk.Frame(self, padding=10, relief='solid', borderwidth=1, style='My.TFrame')
        guess_frame.pack(pady=10)

        guess_label = ttk.Label(guess_frame, text="Entrez votre devinette:", style='My.TLabel')
        guess_label.grid(row=0, column=0, padx=5)
        self.guess_entry = ttk.Entry(guess_frame, textvariable=self.guess_var, width=24, style='My.TEntry')
        self.guess_entry.grid(row=0, column=1, padx=5)

        guess_button = ttk.Button(guess_frame, text="Deviner", command=self.guess, width=10, style='My.TButton')
        guess_button.grid(row=1, column=1, pady=10, padx=5)

        result_frame = ttk.Frame(self, padding=10, relief='solid', borderwidth=1, style='My.TFrame')
        result_frame.pack(pady=10)

        self.result_label = ttk.Label(result_frame, textvariable=self.result_var, style='My.TLabel')
        self.result_label.pack()
        
        self.msg_niveau_frame = ttk.Frame(self, padding=10, relief='solid', borderwidth=1, style='My.TFrame')
        self.msg_niveau_frame.pack(pady=15)
        
        self.msg_niveau = ttk.Label(self.msg_niveau_frame, textvariable=self.msg_niveau_var, style='My.TLabel')
        self.msg_niveau.pack()

    def guess(self):
        """La Fonction utilisée pour déviner le nombre caché"""
        
        guess = int(self.guess_var.get())
        choix = str(self.choixNiveau.get())
        
        if choix == 'Niveau1':
            self.msg_niveau_var.set("Niveau 1: Le nombre caché est compris entre 1 et 100 \net vous avez 60 sécondes pour le trouver")
            fin = time.time()
            temps_ecoule = fin - self.time_start1
            if temps_ecoule <= self.time_limit1:
                if guess == self.secret_number1:
                    reponse = f"Félicitations, vous avez deviné le nombre caché en {int(temps_ecoule)} secondes !"
                    self.result_var.set(reponse)
                    self.secret_number1 = random.randint(1, 100)
                elif guess < self.secret_number1:
                    self.result_var.set("Le nombre caché est plus grand!")
                else:
                    self.result_var.set("Le nombre caché est plus petit!")
                self.guess_entry.delete(0, tk.END)
            else:
                reponse_miss = f"Désolé, vous n'avez pas réussi à deviner le nombre caché en moins de {self.time_limit1} secondes! qui était {int(self.secret_number1)}"
                self.result_var.set(reponse_miss)
                self.secret_number1 = random.randint(1, 100)
                self.guess_entry.delete(0, tk.END)
                
        elif choix == 'Niveau2':
            self.msg_niveau_var.set("Niveau 2: Le nombre caché est compris entre 1 et 250 \net vous avez 45 sécondes pour le trouver")
            fin = time.time()
            temps_ecoule = fin - self.time_start2
            if temps_ecoule <= self.time_limit2: 
                if guess == self.secret_number2:
                    reponse = f"Félicitations, vous avez deviné le nombre caché en {int(temps_ecoule)} secondes !"
                    self.result_var.set(reponse)
                    self.secret_number2 = random.randint(1, 250)
                elif guess < self.secret_number2:
                    self.result_var.set("Le nombre caché est plus grand!")
                else:
                    self.result_var.set("Le nombre caché est plus petit!")
                self.guess_entry.delete(0, tk.END)
            else:
                reponse_miss = f"Désolé, vous n'avez pas réussi à deviner le nombre caché en moins de {self.time_limit2} secondes qui était {int(self.secret_number2)}"
                self.result_var.set(reponse_miss)
                self.secret_number2 = random.randint(1, 250)
                self.guess_entry.delete(0, tk.END)
                
        elif choix == 'Niveau3':
            self.msg_niveau_var.set("Niveau 3: Le nombre caché est compris entre 1 et 1000\net vous avez 30 sécondes pour le trouver")
            fin = time.time()
            temps_ecoule = fin - self.time_start3
            if temps_ecoule <= self.time_limit3: 
                if guess == self.secret_number3:
                    reponse = f"Félicitations, vous avez deviné le nombre caché en {int(temps_ecoule)} secondes !"
                    self.result_var.set(reponse)
                    self.secret_number3 = random.randint(1, 1000)
                elif guess < self.secret_number3:
                    self.result_var.set("Le nombre caché est plus grand!")
                else:
                    self.result_var.set("Le nombre caché est plus petit!")
                self.guess_entry.delete(0, tk.END)
            else:
                reponse_miss = f"Désolé, vous n'avez pas réussi à deviner le nombre caché en moins de {self.time_limit3} secondes qui était {int(self.secret_number3)}"
                self.result_var.set(reponse_miss)
                self.secret_number3 = random.randint(1, 1000)
                self.guess_entry.delete(0, tk.END)
            

if __name__ == '__main__':
    game = GuessNumberGame()
    game.mainloop()
```
